```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
bbc_data = pd.read_csv('bbc_news.csv')
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data['title'])
```

```python
titles.head()
```

## Clean Data

```python
# lowercase
titles['lowercase'] = titles['title'].str.lower()
```

```python
# stop word removal
en_stopwords = stopwords.words('english')
titles['no_stopwords'] = titles['lowercase'].apply(lambda x: ' '.join(word for word in x.split() if word not in (en_stopwords)))
```

```python
# punctuation removal
titles['no_stopwords_no_punct'] = titles.apply(lambda x: re.sub(r'([^\w\s])', '', x['no_stopwords']), axis=1)
```

```python
# tokenize
titles['tokens_raw'] = titles.apply(lambda x: word_tokenize(x['title']), axis=1)
titles['tokens_clean'] = titles.apply(lambda x: word_tokenize(x['no_stopwords_no_punct']), axis=1)
print(titles['tokens_clean'])
```

```python
# lemmatize
lemmatizer = WordNetLemmatizer()
titles['tokens_clean_lemmatized'] = titles['tokens_clean'].apply(
    lambda tokens: [lemmatizer.lemmatize(token) for token in tokens]
)
```

```python
titles.head()
```

```python
# create list for just our tokens
tokens_raw_list = sum(titles['tokens_raw'], []) #unpack our list into a single list
tokens_clean_list = sum(titles['tokens_clean_lemmatized'], [])
```

## POS Tagging

```python
nlp = spacy.load('en_core_web_sm')
```

```python
spacy_doc = nlp(' '.join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(columns = ['token', 'pos_tag'])
```

```python
for token in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{'token': token.text, 'pos_tag': token.pos_}])], ignore_index=True)
```

```python
pos_df_counts = pos_df.groupby(['token', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
pos_df_counts
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == "NOUN"][:10]
nouns
```

```python
verbs = pos_df_counts[pos_df_counts.pos_tag == "VERB"][:10]
verbs
```

```python
nlp = spacy.load('en_core_web_sm')
spacy_doc_clean = nlp(' '.join(tokens_clean_list))
pos_df_clean = pd.DataFrame(columns = ['token', 'pos_tag'])
for token in spacy_doc_clean:
    pos_df_clean = pd.concat([pos_df_clean, pd.DataFrame.from_records([{'token': token.text, 'pos_tag': token.pos_}])], ignore_index=True)
pos_df_counts_clean = pos_df_clean.groupby(['token', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
pos_df_counts_clean
```

```python
nouns = pos_df_counts_clean[pos_df_counts_clean.pos_tag == "NOUN"][:10]
nouns
```

```python
verbs = pos_df_counts_clean[pos_df_counts_clean.pos_tag == "VERB"][:10]
verbs
```
